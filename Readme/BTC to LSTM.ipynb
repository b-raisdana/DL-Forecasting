```python
from FigurePlotter.plotter import timeframe_color

fig = go.Figure()
for timeframe in mt_ohlcva.index.get_level_values(level='timeframe').unique():
    ohlcva = single_timeframe(mt_ohlcva, timeframe)
    fig.add_trace(go.Scatter(
        x=ohlcva.index.get_level_values(level='date'),  # Replace 'time' with your actual time/date column
        y=ohlcva['volume_rma'],
        mode='lines',
        name='Volume RMA',
        line=dict(color=timeframe_color(timeframe))
    ))
    fig.update_layout(
        title='Volume RMA Over Time',
        xaxis_title='Time',
        yaxis_title='Volume RMA',
        template='plotly_white'
    )
    fig.show()
```
