```python
import pandas as pd
import numpy as np
from scipy.stats import randint


def generate_random_dates_scipy(start, end, n_dates=10):
    # Convert start and end to pandas datetime
    start = pd.to_datetime(start)
    end = pd.to_datetime(end)

    # Total number of days between start and end
    days_between = (end - start).days

    # Generate random days using scipy.stats.randint
    random_days = randint.rvs(0, days_between - 10, size=n_dates)
    random_start = start + pd.to_timedelta(random_days, unit='D')

    # Generate random end dates (at least 10 days after start)
    random_end = random_start + pd.to_timedelta(randint.rvs(10, days_between - random_days), unit='D')

    # Return as a DataFrame
    return tuple(zip(random_start, random_end))  # pd.DataFrame({"Start": random_r_start, "End": random_r_end})


# Example usage
start_date = mt_ohlcva.index.get_level_values(level='date').min()
end_date = mt_ohlcva.index.get_level_values(level='date').max()

random_date_pairs = generate_random_dates_scipy(start_date, end_date)
print(random_date_pairs)
```
